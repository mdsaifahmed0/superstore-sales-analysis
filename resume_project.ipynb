```python
category_sales = df.groupby('Category')['Sales'].sum().round(2)

plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().round(2).sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title('Region-wise Sales')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', data=df, alpha=0.5)
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(segment_sales.values, labels=segment_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Segment-wise Sales')
plt.tight_layout()
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Sales'].sum().round(2)

plt.figure(figsize=(8,5))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, marker='o')
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks([2011, 2012, 2013, 2014])
plt.tight_layout()
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Sales'].sum().round(2)

first_year = yearly_sales[2011]
last_year  = yearly_sales[2014]

growth = ((last_year - first_year) / first_year) * 100
print(f"Overall Growth : {growth:.2f}%")
```
